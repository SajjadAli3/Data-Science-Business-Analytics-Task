# Factor contribution of win or loss of a team

```python
Matches = pd.read_csv("IPL/deliveries.csv")
```

```python
Matches.head()
```

```python
# Win or Loss Match 1
Match_1 = Matches[Matches["match_id"]==1]
```

```python
Match_1.head()
```

```python
# inning 1 on Match 1
Match_1_inning_1 = Match_1[Match_1["inning"]==1]
```

```python
Match_1_inning_1.head()
```

```python
Match_1_inning_1.shape
```

```python
# top Scorer
inning_1_batsman_score= Match_1_inning_1[["batsman","batsman_runs"]].groupby("batsman").sum().sort_values(by="batsman_runs",ascending=False)
```

```python
inning_1_batsman_score
```

```python
inning_1_batsman_score.plot(kind='bar',color='lightgreen',figsize=(15,5))
plt.xticks(rotation=90)
plt.show()
```

```python
#over Score
Inning_1_over_score=Match_1_inning_1[["over","total_runs"]].groupby("over").sum()
Inning_1_over_score.head(5)
```

```python
Inning_1_over_score.plot(kind='bar',color='purple',figsize=(15,5))
plt.title("1st Inning Score by overs")
plt.xlabel("Overs",fontsize=30)
plt.show()
```

```python
# inning 1 on Match 1
Match_2_inning_2 = Match_1[Match_1["inning"]==2]
```

```python
# top Scorer
inning_2_batsman_score= Match_2_inning_2[["batsman","batsman_runs"]].groupby("batsman").sum().sort_values(by="batsman_runs",ascending=False)
```

```python
inning_2_batsman_score
```

```python
inning_2_batsman_score.plot(kind='bar',color='black',figsize=(15,5))
plt.xticks(rotation=90)
plt.show()
```

```python
#2nd inning over Scores
Inning_2_over_score=Match_2_inning_2[["over","total_runs"]].groupby("over").sum()
Inning_2_over_score.head(5)
```

```python
#bar plot for Overs
Inning_2_over_score.plot(kind='bar',color='purple',figsize=(15,5))
plt.title("2nd Inning Score by overs")
plt.xlabel("Overs",fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(Match_1_inning_1["over"].value_counts().keys(),Inning_1_over_score['total_runs'],label='Inning 1')
sns.scatterplot(Match_1_inning_1["over"].value_counts().keys(),Inning_1_over_score['total_runs'])
sns.lineplot(Match_2_inning_2["over"].value_counts().keys(),Inning_2_over_score['total_runs'],label='Inning 2')
sns.scatterplot(Match_2_inning_2["over"].value_counts().keys(),Inning_2_over_score['total_runs'])
plt.show()
#Inning_1_over_score.plot(kind='line',figsize=(15,5))
#Inning_2_over_score.plot(kind='line',color='red',figsize=(15,5))
#plt.show()
```

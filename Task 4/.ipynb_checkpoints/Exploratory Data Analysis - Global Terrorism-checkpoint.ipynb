```python
#import required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load Dataset
Dataset = pd.read_csv("globalterrorismdb_0718dist.csv")
```

```python
Dataset.head()
```

```python
Dataset.shape
```

```python
Dataset.columns
```

```python
Dataset['iyear'].isnull().sum()
```

```python
#Drop columns Which null values > 45%
for col in Dataset.columns:
    if (Dataset[col].isnull().sum()/len(Dataset.index)*100)>  45:
        Dataset.drop(col,axis=1,inplace=True)
```

```python
Dataset.shape
```

```python
Dataset.head()
```

```python
#Drop columns Which have 0 0r 0.0 values > 65%
for col in Dataset.columns:
    d = Dataset[col].values == (0 or 0.0)
    if (d.sum()/len(Dataset.index)*100)>65:
        Dataset.drop(col,axis=1,inplace=True)
```

```python
Dataset.shape
```

```python
Dataset.columns
```

```python
#Counts of Attack Type
plt.figure(figsize=(15,5))
p=sns.countplot('attacktype1_txt',data = Dataset)
p.set_xticklabels(p.get_xticklabels(),rotation=90)
plt.show()
```

```python
Dataset.head()
```

```python
#rename Dataset Columns
Dataset.rename(columns={
    "iyear":"Year",
    "imonth":"Month",
    "iday":"Day",
    "country_txt":"Country",
    "region_txt":"Region",
    "provstate":"Province/State",
    "city":"City",
    "latitude":"Latitude",
    "longitude":"Longitude",
    "summary":"Summary",
    "attacktype1_txt":"Attack Type",
    "targtype1_txt":"Target Type",
    "gname":"Group Name",
    "weaptype1_txt":"Weapon Type",
    "nkill":"Killed",
    "nwound":"Wounded",             
},inplace=True)
```

```python
Killed_by_region=Dataset[["Region","Killed"]].groupby("Region").sum().sort_values(by='Killed',ascending=False)
```

```python
Killed_by_region
```

```python
Killed_by_region.plot(kind="bar",color="purple",figsize=(10,5))
plt.title("People Killed in each Region")
plt.xlabel("Regions",fontsize=12)
plt.ylabel("Number of People Killed",fontsize=12) 
plt.show()
```

```python
Killed_by_Country_Group =Dataset[['Group Name','Country','Killed']].groupby(['Group Name','Country']).sum().sort_values(by='Killed',ascending=False).drop('Unknown').head(10)
Killed_by_Country_Group
```

```python
# Countries with most people killed by a particular Group
Killed_by_Country_Group.plot(kind='bar',color='lightblue',figsize=(15,10))
plt.title("Countries with most people killed by a particular Group",fontsize=13)
plt.xlabel("(Terrorist Group , Country)",fontsize=13)
plt.xticks(fontsize=10)
plt.ylabel("Number of people Killed",fontsize=13) 
plt.show()
```

# Prediction Using Decision Tree Algorithm

```python
import pandas as pd
```

```python
Dataset = pd.read_csv("Iris.csv")
Dataset.head()
```

# Set Dependent & Independent Columns

```python
input_value  = Dataset.drop(["Id","Species"],axis=1)
Target_value =Dataset[["Species"]]
```

```python
Target_value
```

# Convert Target values to Numerical Values

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
Target_value["species_n"]=LE.fit_transform(Target_value)
```

```python
Target_value
```

```python
Target_value.drop("Species",axis='columns',inplace=True)
```

```python
Target_value.head()
```

# Import Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT = DecisionTreeClassifier()
```

```python
# fit the Values for prediction
DT.fit(input_value,Target_value)
```

```python
DT.score(input_value,Target_value)
```

```python
DT.predict([[3,3.9,1.7,0.1]])
```

# Now Visualize the Decision Tree for better understanding

```python
# Import necessary libraries for graph viz
from six import StringIO
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

```python
# Visualize the graph
d_data = StringIO()
export_graphviz(DT, out_file=d_data, feature_names=input_value.columns,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(d_data.getvalue())  
Image(graph.create_png())
```
